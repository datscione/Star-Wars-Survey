# Remove rows where RespondentID is NaN

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
```

```python
star_wars.head(10)
```

# Change Yes/No columns to Boolean True/False columns

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
# Use a dictionary to define a mapping for Yes and No
yes_no = {'Yes': True, 'No': False}
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
import numpy as np

seen_film = {
    "Star Wars: Episode I  The Phantom Menace":True, 
    np.nan:False,
    "Star Wars: Episode II  Attack of the Clones":True, 
    "Star Wars: Episode III  Revenge of the Sith":True, 
    "Star Wars: Episode IV  A New Hope":True, 
    "Star Wars: Episode V The Empire Strikes Back":True, 
    "Star Wars: Episode VI Return of the Jedi":True}

for col in star_wars.columns[3:9]:
        star_wars[col] = star_wars[col].map(seen_film)
```

```python
star_wars.head(10)
```

```python
star_wars.rename(columns=
                 {'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
                 'Unnamed: 4':'seen_2','Unnamed: 5':'seen_3', 'Unnamed: 6':'seen_4', 
                 'Unnamed: 7':'seen_5', 'Unnamed: 8':'seen_6'}, inplace=True)
```

```python
star_wars.head(10)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars[star_wars.columns[9:15]]
```

```python
star_wars.rename(columns=
                 {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
                 'Unnamed: 10':'ranking_2','Unnamed: 11':'ranking_3', 'Unnamed: 12':'ranking_4', 
                 'Unnamed: 13':'ranking_5', 'Unnamed: 14':'ranking_6'}, inplace=True)
```

```python
star_wars[star_wars.columns[9:15]]
```

# Find highest ranking movie.

```python
star_wars[star_wars.columns[9:15]].mean()
```

The highest ranking film is the fifth one, "Star Wars: Episode V The Empire Strikes Back".

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
plt.xlabel("Star Wars Films 1-6")
plt.ylabel("Mean Rankings")
```

The fifth film, "Star Wars: Episode V The Empire Strikes Back" had the lowest mean ranking. It was the closest to 1, the score for the favorite episode, making it the most popular of the six films.

# Determine how many people saw each movie.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
%matplotlib inline
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
plt.xlabel("Star Wars Films 1-6")
plt.ylabel("Number of Viewers")
```

The fith film, the most popular one, is also the Star Wars movie that had the most number of views. These agree with it being the favorite film.

```python
fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]
not_a_fan = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == False]
```

```python
fan.sum()
```

```python
fan.mean()
```

Above are population and mean value dataframes for Star Wars fans for each column. 

```python
%matplotlib inline
plt.bar(range(6), fan[fan.columns[3:9]].sum())
plt.xlabel("Star Wars Films 1-6")
plt.ylabel("Number of Viewers")
```

```python
plt.bar(range(6), fan[fan.columns[9:15]].mean())
plt.xlabel("Star Wars Films 1-6")
plt.ylabel("Mean Rankings")
```

Among fans, the most viewed movies were episodes five and six. The fifth episode was generally their favorite one. Let's look at the results for non-fans.

```python
plt.bar(range(6), not_a_fan[not_a_fan.columns[3:9]].sum())
plt.xlabel("Star Wars Films 1-6")
plt.ylabel("Number of Viewers")
```

```python
plt.bar(range(6), not_a_fan[not_a_fan.columns[9:15]].mean())
plt.xlabel("Star Wars Films 1-6")
plt.ylabel("Mean Rankings")
```

So we see that individuals who are not Star Wars fans watched the fifth episode more than any other episode.

```python
not_a_fan.mean()
```

Interestingly, the favorite episode among non-fans was the sixth Star Wars episode.

In conclusion, "Star Wars: Episode V The Empire Strikes Back" was the most popular and viewed of the Star Wars movies. This was the case among fans and non-fans alike.
